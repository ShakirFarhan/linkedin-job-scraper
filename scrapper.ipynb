```python
from selenium import webdriver
import time
import pandas as pd
import os
#Import Packages

from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

```

```python
jobsUrl='https://www.linkedin.com/jobs/search/?currentJobId=3802067798&distance=25&f_E=1%2C2&geoId=105556991&keywords=react&origin=JOBS_HOME_SEARCH_CARDS'
driver=webdriver.Chrome()
driver.get(jobsUrl)
wait = WebDriverWait(driver, 10)
```

```python
job_count = driver.find_element(By.CLASS_NAME, 'results-context-header__job-count').text
# Extract the text from the found element

nJob_count=n=pd.to_numeric(job_count)
nJob_count
```

```python
i=2
while(i<=30):
  driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
  i=i+1
  try:
    see_more_btn=driver.find_element(By.XPATH,"//button[@aria-label='See more jobs']")
    driver.execute_script("arguments[0].click();", see_more_btn)   
    time.sleep(3)
  except:
    pass
    time.sleep(3)
```

```python
companyNames=[]
companyElements = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, 'base-search-card__subtitle')))
```

```python
try:
  for companyElement in companyElements:
    try:
        hiddenLink = companyElement.find_element(By.CLASS_NAME, 'hidden-nested-link')
        companyName = hiddenLink.text.strip()
    except:
        companyName = companyElement.text.strip()
    companyNames.append(companyName)
except IndexError:
    print("Completed")
```

```python
companyNames
```

```python
jobTitles = []
titles = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, 'base-search-card__title')))
try:
  for i, title in enumerate(titles):
    jobTitles.append(title.text)
except IndexError:
  print("Completed")
```

```python
jobTitles
```

```python
jobLinks=[]
jobLinksElements = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, "base-card__full-link")))
for link in jobLinksElements:
  link = link.get_attribute("href")
  jobLinks.append(link)
```

```python
jobLinks
```

```python
jobLocations=[]
jobLocationElements = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, "job-search-card__location")))
for location in jobLocationElements:
  link = location.text
  jobLocations.append(link)
```

```python
jobLocations
```

```python
companyNameFinal=pd.DataFrame(companyNames,columns=["company"])
jobTitleFinal=pd.DataFrame(jobTitles,columns=["title"])
jobLinks=pd.DataFrame(jobLinks,columns=["link"])
jobLocations=pd.DataFrame(jobLocations,columns=["location"])
job=companyNameFinal.join(jobTitleFinal).join(jobLocations)
```
